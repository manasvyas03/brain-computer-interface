```python
import numpy as np
import mne
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Define the list of file paths to your EDF files
file_paths = ['EEG dataset\h01.edf', 'EEG dataset\h02.edf', 'EEG dataset\h03.edf', 'EEG dataset\h04.edf', 'EEG dataset\h05.edf', 'EEG dataset\h06.edf', 'EEG dataset\h07.edf', 'EEG dataset\h08.edf', 'EEG dataset\h09.edf', 'EEG dataset\h10.edf', 'EEG dataset\h11.edf', 'EEG dataset\h12.edf', 'EEG dataset\h13.edf', 'EEG dataset\h14.edf']  # Replace with your actual file paths

# Initialize lists to store EEG data and labels
X = []
y = []

# Define the maximum length of EEG samples
max_length = 0

# Iterate through each EDF file
for file_path in file_paths:
    # Load EDF file
    raw = mne.io.read_raw_edf(file_path, preload=True)
    
    # Preprocess EEG data if needed (filtering, normalization, etc.)
    # Example: raw.filter(0.5, 30)
    
    # Extract EEG data and label from file
    eeg_data = raw.get_data()  # Assuming all channels are EEG channels
    label = file_path.split('/')[-1].split('.')[0]  # Extract label from file name
    
    # Update maximum length
    max_length = max(max_length, eeg_data.shape[1])
    
    # Append data and label to lists
    X.append(eeg_data)
    y.append(label)

# Pad or trim EEG data to ensure uniform shape
X_padded = pad_sequences(X, maxlen=max_length, dtype='float32', padding='post', truncating='post')

# Convert lists to numpy arrays
X_padded = np.array(X_padded)
y = np.array(y)

# Define RNN model architecture
model = Sequential([
    LSTM(units=64, input_shape=(X_padded.shape[1], X_padded.shape[2])),
    Dropout(0.5),
    Dense(units=128, activation='relu'),
    Dense(units=len(np.unique(y)), activation='softmax')  # Output layer with number of classes
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the RNN model
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
import numpy as np
import mne
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Define the list of file paths to your EDF files
file_paths = ['EEG dataset/s01.edf', 'EEG dataset/s02.edf', 'EEG dataset/s03.edf', 'EEG dataset/s04.edf', 'EEG dataset/s05.edf', 'EEG dataset/s06.edf', 'EEG dataset/s07.edf', 'EEG dataset/s08.edf', 'EEG dataset/s09.edf', 'EEG dataset/s10.edf', 'EEG dataset/s11.edf', 'EEG dataset/s12.edf', 'EEG dataset/s13.edf', 'EEG dataset/s14.edf']  # Replace with your actual file paths


# Initialize lists to store EEG data and labels
X = []
y = []

# Iterate through each EDF file
for file_path in file_paths:
    # Load EDF file
    raw = mne.io.read_raw_edf(file_path, preload=True)
    
    # Preprocess EEG data if needed (filtering, normalization, etc.)
    # Example: raw.filter(0.5, 30)
    
    # Extract EEG data and label from file
    eeg_data = raw.get_data()  # Assuming all channels are EEG channels
    
    # Ensure consistent shape of EEG data
    # If the shape is inconsistent, consider padding or trimming to a fixed length
    max_length = 1000  # Example: set a maximum length
    if eeg_data.shape[1] < max_length:
        # Pad the data to the maximum length
        eeg_data = np.pad(eeg_data, ((0, 0), (0, max_length - eeg_data.shape[1])), mode='constant')
    elif eeg_data.shape[1] > max_length:
        # Trim the data to the maximum length
        eeg_data = eeg_data[:, :max_length]
    
    # Append data and label to lists
    X.append(eeg_data)
    y.append(file_path.split('/')[-1].split('.')[0])  # Extract label from file name

# Convert lists to numpy arrays
X = np.array(X)
y = np.array(y)

# Flatten the data (convert 2D array to 1D array)
X_flat = X.reshape(X.shape[0], -1)

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'gamma': ['scale', 'auto', 1, 0.1, 0.01],  # Kernel coefficient
    'kernel': ['linear', 'rbf', 'poly']  # Kernel type
}

# Initialize SVM classifier
svm = SVC()

# Use StratifiedKFold with a smaller number of splits
stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# Initialize GridSearchCV with StratifiedKFold
grid_search = GridSearchCV(svm, param_grid, cv=stratified_kfold)

# Train the SVM classifier with hyperparameter tuning
grid_search.fit(X_flat, y)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Calculate accuracy using cross-validation
cross_val_accuracy = grid_search.best_score_
print("Cross-Validation Accuracy:", cross_val_accuracy)
```
